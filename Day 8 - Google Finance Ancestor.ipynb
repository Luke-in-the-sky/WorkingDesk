```python
import requests
import ast

stock = 'SPY'

api_url = 'https://www.quandl.com/api/v3/datasets/WIKI/%s/data.json' % stock
session = requests.Session()
session.mount('http://', requests.adapters.HTTPAdapter(max_retries=3))
raw_data = session.get(api_url)
```

```python
import pprint as pp


raw_data_js = raw_data.json()['dataset_data']
raw_data_js.keys()
#data_d
```

```python
import pandas as pd


data = pd.DataFrame(raw_data_js['data'], columns=raw_data_js['column_names'])
data.head()
```

```python
from bokeh.charts import output_notebook, show
output_notebook()
```

```python
from datetime import datetime

data['DateTime'] = data['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d')  )
```

```python
data[['Date', 'DateTime']].head()
```

```python
from datetime import datetime
from bokeh.plotting import figure
#TOOLS = 'pan,wheel_zoom'

#  datetime.strptime(self.request.get('date'), '%Y-%m-%d')

x = data['DateTime']
y = data['Close']
p = figure(title="Quandle data",
           x_axis_label = "Date", 
           x_axis_type = "datetime")
p.line(x=x, y=y, line_color='red', legend="Close price")
p.line(x=x, y=data['Open'], line_color='navy', legend='Open price')
#output_file("scatter.html")

p.legend.orientation = "top_left"

show(p)
```

```python
from bokeh.embed import components

script, div = components(p)
div
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_file

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

output_file("legend_labels.html")

p = figure()

p.circle(x, y, legend="sin(x)")
p.line(x, y, legend="sin(x)")

p.line(x, 2*y, legend="2*sin(x)",
       line_dash=[4, 4], line_color="orange", line_width=2)

p.square(x, 3*y, legend="3*sin(x)", fill_color=None, line_color="green")
p.line(x, 3*y, legend="3*sin(x)", line_color="green")

p.legend.location= "bottom_left"

show(p)
```
